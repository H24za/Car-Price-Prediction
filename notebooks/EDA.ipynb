```python
def top_and_bottom_combined(df, col, n=5):
    top_counts = df[col].value_counts().head(n)
    bottom_counts = df[col].value_counts().tail(n)
    
    combined = pd.concat([top_counts, bottom_counts], axis=0)
    return combined

for col in data.select_dtypes(include='object'):
    print(f'{col} value counts:')
    print(top_and_bottom_combined(data, col))
    print('30 Unique values:', data[col].unique()[:30])
    print('-------------------')
```

```python
# Fix the problems
# Replace Levy '-' with 0 and convert column type safely (avoid inplace on slice)
# Use assignment to the original DataFrame to avoid pandas chained-assignment future warning
data['Levy'] = data['Levy'].replace({'-': 0})
# Convert to numeric, coercing errors to NaN, then fill NaN with 0 (or decide appropriate strategy)
data['Levy'] = pd.to_numeric(data['Levy'], errors='coerce').fillna(0)
```

```python
data['Engine volume'] = data['Engine volume'].str.replace('Turbo', '')
data['Engine volume'] = pd.to_numeric(data['Engine volume'])
```

```python
data['Mileage'] = data['Mileage'].str.replace('km', '')
data['Mileage'] = pd.to_numeric(data['Mileage'])
```

```python
data.drop(['Doors'], axis=1, inplace=True)
```

### 4. Correlations
- Numerical Correlations (`corr()`)
- Categorical Correlations

```python
# Numeric-only correlation heatmap (avoid trying to convert strings to float)
num_cols = data.select_dtypes(include=['number']).columns
if len(num_cols) == 0:
    print('No numeric columns available for correlation matrix.')
else:
    corr = data[num_cols].corr()
    # mask upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlation matrix (numeric features)')
    plt.show()
```
